```python
data = Table.read_table('file:///Users/ishandikshit/Desktop/bombassbitchv2.csv')
```

```python
data = data.sort('sum', descending = True)
```

```python
data = data.drop('2012-08')
data = data.drop('2012-01')
data
```

```python
data = data.drop('sum')
data
```

```python
dates = [ass for ass in data.labels]
dates = dates[1:]
transpose = Table().with_column('dates', dates)
transpose
```

```python
for index in np.arange(data.num_rows):
    transpose.append_column(str(np.array(data.row(index))[0]), np.array(data.row(index))[1:])
```

```python
transpose = transpose.sort('dates')
```

```python
transpose
```

```python
tab = Table.read_table('file:///Users/ishandikshit/Desktop/bizseven.csv')
dicty = {}
tab = tab.select([2,3])

for i in range(tab.num_rows):
   dicty[tab.take(i).column(1)[0]] = tab.take(i).column(0)[0]
```

```python
for i in np.arange(1,len(transpose.labels)):
    transpose = transpose.relabel(transpose.labels[i], dicty[int(transpose.labels[i][:-2])])
```
